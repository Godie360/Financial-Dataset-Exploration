```python
for index, row in dataset.iterrows():
    duplicate_count = row.duplicated().sum()
    if duplicate_count > 0:
        duplicates = row[row.duplicated()]
        print(f"Row {index + 1}:")
        print(f"Duplicate Count: {duplicate_count}")
        print("Duplicate Values:")
        print(duplicates)
        print()
```
